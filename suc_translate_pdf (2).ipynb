```python
import fitz  # PyMuPDF
import arabic_reshaper
from bidi.algorithm import get_display
import os

def correct_arabic_text_in_pdf(input_pdf_path, output_pdf_path):
    """
    يقرأ ملف PDF ويصحح النص العربي المعكوس في كل صفحة.

    :param input_pdf_path: المسار إلى ملف PDF الأصلي.
    :param output_pdf_path: المسار لحفظ ملف PDF المصحح.
    """
    try:
        doc = fitz.open(input_pdf_path)
        new_doc = fitz.open()

        for page_num in range(len(doc)):
            page = doc.load_page(page_num)
            new_page = new_doc.new_page(width=page.rect.width, height=page.rect.height)

            # استخراج النصوص مع معلومات التنسيق والموقع
            blocks = page.get_text("dict")["blocks"]
            for b in blocks:
                if "lines" in b:
                    for l in b["lines"]:
                        for s in l["spans"]:
                            text = s["text"]
                            # افتراض أن معظم النص هو باللغة العربية
                            # إعادة تشكيل النص العربي وتصحيح اتجاهه
                            reshaped_text = arabic_reshaper.reshape(text)
                            bidi_text = get_display(reshaped_text)

                            # الحصول على خصائص النص الأصلي
                            fontname = s["font"]
                            fontsize = s["size"]
                            color_srgb = tuple(c for c in s["color"])
                            origin = (s["origin"][0], s["origin"][1] - 1)  # تعديل طفيف للموقع العمودي

                            # إدراج النص المصحح في الصفحة الجديدة
                            new_page.insert_text(origin,
                                                  bidi_text,
                                                  fontname=fontname,
                                                  fontsize=fontsize,
                                                  color=color_srgb,
                                                  )

        # حفظ الملف الجديد
        new_doc.save(output_pdf_path)
        print(f"تم حفظ الملف المصحح في: {output_pdf_path}")

    except Exception as e:
        print(f"حدث خطأ: {e}")

# --- كيفية الاستخدام ---
if __name__ == "__main__":
    # ضع هنا مسار ملف الإدخال الذي تم إنشاؤه بواسطة pdf2zh
    input_file = "/content/PDFMathTranslate/translate.cli.text.with.figure-dual.pdf"

    # التأكد من وجود ملف الإدخال
    if not os.path.exists(input_file):
        print(f"خطأ: ملف الإدخال '{input_file}' غير موجود.")
    else:
        # تحديد اسم ملف الإخراج
        output_file = os.path.splitext(input_file)[0] + "_corrected.pdf"
        correct_arabic_text_in_pdf(input_file, output_file)
```

```python
import fitz  # PyMuPDF
import arabic_reshaper
from bidi.algorithm import get_display
import os

def correct_arabic_text_in_pdf(input_pdf_path, output_pdf_path):
    """
    يقرأ ملف PDF ويصحح النص العربي المعكوس في كل صفحة.

    :param input_pdf_path: المسار إلى ملف PDF الأصلي.
    :param output_pdf_path: المسار لحفظ ملف PDF المصحح.
    """
    try:
        doc = fitz.open(input_pdf_path)
        new_doc = fitz.open()

        for page_num in range(len(doc)):
            page = doc.load_page(page_num)
            new_page = new_doc.new_page(width=page.rect.width, height=page.rect.height)

            # استخراج النصوص مع معلومات التنسيق والموقع
            blocks = page.get_text("dict")["blocks"]
            for b in blocks:
                if b.get("type") == 0 and "lines" in b:  # التأكد من أنه بلوك نصي
                    for l in b["lines"]:
                        for s in l["spans"]:
                            text = s["text"]

                            # إعادة تشكيل النص العربي وتصحيح اتجاهه
                            reshaped_text = arabic_reshaper.reshape(text)
                            bidi_text = get_display(reshaped_text)

                            # --- بداية التعديل ---
                            # الحصول على خصائص النص الأصلي
                            fontname = s["font"]
                            fontsize = s["size"]
                            origin = s["origin"]

                            # معالجة اللون بشكل صحيح
                            # s["color"] هو رقم صحيح، يجب تحويله إلى tuple (r, g, b)
                            color_int = s["color"]
                            r = ((color_int >> 16) & 0xFF) / 255.0
                            g = ((color_int >> 8) & 0xFF) / 255.0
                            b = (color_int & 0xFF) / 255.0
                            color_rgb = (r, g, b)
                            # --- نهاية التعديل ---

                            # إدراج النص المصحح في الصفحة الجديدة
                            # نستخدم insert_text بدلاً من textbox للحفاظ على الموقع الدقيق للحرف الأول
                            new_page.insert_text(origin,
                                                  bidi_text,
                                                  fontname=fontname,
                                                  fontsize=fontsize,
                                                  color=color_rgb
                                                  )

        # حفظ الملف الجديد
        new_doc.save(output_pdf_path, garbage=4, deflate=True)
        new_doc.close()
        doc.close()
        print(f"تم حفظ الملف المصحح في: {output_pdf_path}")

    except Exception as e:
        print(f"حدث خطأ: {e}")

# --- كيفية الاستخدام ---
if __name__ == "__main__":
    # ضع هنا مسار ملف الإدخال الذي تم إنشاؤه بواسطة pdf2zh
    input_file = "/content/PDFMathTranslate/translate.cli.text.with.figure-dual.pdf"

    # التأكد من وجود ملف الإدخال
    if not os.path.exists(input_file):
        print(f"خطأ: ملف الإدخال '{input_file}' غير موجود.")
    else:
        # تحديد اسم ملف الإخراج
        output_file = os.path.splitext(input_file)[0] + "_corrected.pdf"
        correct_arabic_text_in_pdf(input_file, output_file)
```

```python
# ==============================================================================
# 🌟 الكود المُعدّل لحل مشكلة 'need font file or buffer' في Colab
# ==============================================================================

# الخطوة 1: تثبيت المكتبات وحزمة الخطوط العربية الضرورية
# تثبيت حزمة الخطوط Noto Naskh Arabic في بيئة Colab
!pip install PyMuPDF arabic-reshaper python-bidi -q
!apt-get install -y fonts-noto-naskh-arabic -qq

# الخطوة 2: استيراد المكتبات
import fitz  # PyMuPDF
import arabic_reshaper
from bidi.algorithm import get_display
import os

# الخطوة 3: تعريف المسار الثابت للخط العربي
# هذا هو المسار القياسي للخط في بيئة Colab/Debian بعد تثبيته
ARABIC_FONT_PATH = "/usr/share/fonts/truetype/noto/NotoNaskhArabic-Regular.ttf"

# الخطوة 4: تعريف الدالة التي تقوم بتصحيح ملف PDF
def correct_arabic_text_in_pdf(input_pdf_path, output_pdf_path):
    """
    يقرأ ملف PDF ويصحح النص العربي المعكوس باستخدام خط عربي قياسي.
    """
    # التحقق من وجود ملف الخط قبل البدء
    if not os.path.exists(ARABIC_FONT_PATH):
        print(f"❌ خطأ فادح: ملف الخط العربي غير موجود في المسار: {ARABIC_FONT_PATH}")
        print("يرجى التأكد من تشغيل الخلية بنجاح، وخاصة الأمر !apt-get install.")
        return

    try:
        doc = fitz.open(input_pdf_path)
        new_doc = fitz.open()

        print(f"بدء معالجة الملف: {input_pdf_path}")
        print(f"عدد الصفحات: {len(doc)}")

        for page_num in range(len(doc)):
            page = doc.load_page(page_num)
            new_page = new_doc.new_page(width=page.rect.width, height=page.rect.height)

            blocks = page.get_text("dict")["blocks"]
            for b in blocks:
                if b.get("type") == 0 and "lines" in b:
                    for l in b["lines"]:
                        for s in l["spans"]:
                            text = s["text"]

                            # إعادة تشكيل النص العربي وتصحيح اتجاهه
                            reshaped_text = arabic_reshaper.reshape(text)
                            bidi_text = get_display(reshaped_text)

                            # الحصول على خصائص النص الأصلي
                            fontsize = s["size"]
                            origin = s["origin"]

                            # معالجة اللون
                            color_int = s["color"]
                            r = ((color_int >> 16) & 0xFF) / 255.0
                            g = ((color_int >> 8) & 0xFF) / 255.0
                            b = (color_int & 0xFF) / 255.0
                            color_rgb = (r, g, b)

                            # ⭐️⭐️⭐️ التعديل الذي يحل المشكلة ⭐️⭐️⭐️
                            # استخدام 'fontfile' مع المسار المباشر للخط
                            new_page.insert_text(origin,
                                                  bidi_text,
                                                  fontfile=ARABIC_FONT_PATH,
                                                  fontsize=fontsize,
                                                  color=color_rgb
                                                  )

        new_doc.save(output_pdf_path, garbage=4, deflate=True)
        print("-" * 20)
        print(f"✅ تم الانتهاء بنجاح!")
        print(f"تم حفظ الملف المصحح في: {output_pdf_path}")

    except Exception as e:
        print(f"❌ حدث خطأ: {e}")
    finally:
        if 'doc' in locals() and not doc.is_closed:
            doc.close()
        if 'new_doc' in locals() and not new_doc.is_closed:
            new_doc.close()

# --- كيفية الاستخدام ---
# الخطوة 5: تحديد مسار الملفات وتشغيل الدالة

# ==============================================================================
# ⚠️ هام: قم بتعديل هذا السطر ليشير إلى مسار ملف PDF الذي تريد تصحيحه
input_file = "/content/PDFMathTranslate/translate.cli.text.with.figure-dual.pdf"
# ==============================================================================

if not os.path.exists(input_file):
    print(f"خطأ: ملف الإدخال '{input_file}' غير موجود.")
    print("يرجى التأكد من تحميل الملف إلى Colab ووضع المسار الصحيح.")
else:
    output_file = os.path.splitext(input_file)[0] + "_corrected.pdf"
    correct_arabic_text_in_pdf(input_file, output_file)
```

```python
# الخطوة 1: تثبيت المكتبات المطلوبة، بما في ذلك Tesseract OCR
!pip install PyMuPDF arabic-reshaper python-bidi pytesseract Pillow -q
!apt-get install -y tesseract-ocr tesseract-ocr-ara -qq

# الخطوة 2: استيراد المكتبات
import fitz  # PyMuPDF
import arabic_reshaper
from bidi.algorithm import get_display
import os
import pytesseract
from PIL import Image
import io

# الخطوة 3: تعريف المسار لملف الخط العربي
# سنستخدم الخط الذي تم تنزيله سابقًا
!wget -q --show-progress -O NotoNaskhArabic-Regular.ttf "https://github.com/google/fonts/raw/main/ofl/notonaskharabic/NotoNaskhArabic-Regular.ttf"
ARABIC_FONT_PATH = "NotoNaskhArabic-Regular.ttf"

# الخطوة 4: تعريف الدالة الشاملة التي تستخدم OCR
def fix_pdf_with_ocr(input_pdf_path, output_pdf_path, font_path):
    if not os.path.exists(font_path):
        print(f"❌ خطأ فادح: ملف الخط العربي '{font_path}' غير موجود.")
        return

    try:
        doc = fitz.open(input_pdf_path)
        new_doc = fitz.open()

        print(f"\nبدء معالجة الملف باستخدام OCR: {os.path.basename(input_pdf_path)}")
        print(f"عدد الصفحات: {len(doc)}")

        for page_num in range(len(doc)):
            print(f"\n--- معالجة صفحة {page_num + 1} ---")
            page = doc.load_page(page_num)

            # إنشاء صفحة جديدة بنفس الحجم
            new_page = new_doc.new_page(width=page.rect.width, height=page.rect.height)

            # تحويل الصفحة إلى صورة عالية الدقة
            # DPI (Dots Per Inch) - كلما زاد الرقم زادت الدقة
            pix = page.get_pixmap(dpi=300)
            img_bytes = pix.tobytes("png")
            pil_img = Image.open(io.BytesIO(img_bytes))

            # استخدام Tesseract OCR لاستخراج النص العربي
            print("جاري استخراج النص من الصورة...")
            # نستخدم 'ara' لتحديد اللغة العربية
            extracted_text = pytesseract.image_to_string(pil_img, lang='ara')

            if not extracted_text.strip():
                print("لم يتم العثور على أي نص في هذه الصفحة.")
                continue

            print("تم استخراج النص. جاري تصحيحه وكتابته في الملف الجديد...")

            # إصلاح النص المستخرج (OCR قد لا يحتاج لهذا، لكن للتأكيد)
            reshaped_text = arabic_reshaper.reshape(extracted_text)
            bidi_text = get_display(reshaped_text)

            # إضافة النص المصحح إلى الصفحة الجديدة
            # نستخدم textbox لوضع كل النص في مربع واحد
            rect = fitz.Rect(50, 50, page.rect.width - 50, page.rect.height - 50) # هوامش 50 بكسل
            new_page.insert_textbox(rect, bidi_text, fontfile=font_path, fontsize=12, align=fitz.TEXT_ALIGN_RIGHT)


        if len(new_doc) > 0:
            new_doc.save(output_pdf_path, garbage=4, deflate=True)
            print("\n" + "="*20)
            print(f"✅ تم الانتهاء بنجاح!")
            print(f"تم حفظ الملف المصحح في: {output_pdf_path}")
        else:
            print("\n⚠️ تحذير: لم يتم إنشاء ملف جديد لأنه لم يتم العثور على أي نصوص في الملف الأصلي.")

    except Exception as e:
        print(f"❌ حدث خطأ أثناء المعالجة: {e}")
    finally:
        if 'doc' in locals() and not doc.is_closed:
            doc.close()
        if 'new_doc' in locals() and not new_doc.is_closed:
            new_doc.close()

# --- كيفية الاستخدام ---
# الخطوة 6: تحديد مسار الملفات وتشغيل الدالة

# ==============================================================================
# ⚠️ هام: قم بتعديل هذا السطر ليشير إلى مسار ملف PDF الذي تريد تصحيحه
# سواء كان يحتوي على نص معكوس أو مربعات
input_file = "/content/PDFMathTranslate/translate.cli.text.with.figure-dual.pdf"
# ==============================================================================

if not os.path.exists(input_file):
    print(f"خطأ: ملف الإدخال '{input_file}' غير موجود.")
    print("يرجى التأكد من تحميل الملف إلى Colab ووضع المسار الصحيح.")
else:
    output_file = os.path.splitext(input_file)[0] + "_corrected_with_ocr.pdf"
    fix_pdf_with_ocr(input_file, output_file, ARABIC_FONT_PATH)
```

```python
# الخطوة 1: تثبيت المكتبات اللازمة، بما في ذلك Tesseract OCR
!pip install PyMuPDF arabic-reshaper python-bidi pytesseract Pillow -q
!apt-get install -y tesseract-ocr tesseract-ocr-ara -qq

# الخطوة 2: استيراد المكتبات
import fitz  # PyMuPDF
import arabic_reshaper
from bidi.algorithm import get_display
import os
import pytesseract
from PIL import Image
import io

# الخطوة 3: تعريف المسار لملف الخط العربي
!wget -q --show-progress -O NotoNaskhArabic-Regular.ttf "https://github.com/google/fonts/raw/main/ofl/notonaskharabic/NotoNaskhArabic-Regular.ttf"
ARABIC_FONT_PATH = "NotoNaskhArabic-Regular.ttf"

# الخطوة 4: تعريف الدالة الشاملة التي تستخدم OCR
def fix_pdf_with_ocr(input_pdf_path, output_pdf_path, font_path):
    if not os.path.exists(font_path):
        print(f"❌ خطأ فادح: ملف الخط العربي '{font_path}' غير موجود.")
        return

    try:
        doc = fitz.open(input_pdf_path)
        new_doc = fitz.open()

        print(f"\nبدء معالجة الملف باستخدام OCR: {os.path.basename(input_pdf_path)}")
        print(f"عدد الصفحات: {len(doc)}")

        for page_num in range(len(doc)):
            print(f"\n--- معالجة صفحة {page_num + 1} ---")
            page = doc.load_page(page_num)

            # إنشاء صفحة جديدة بنفس الحجم
            new_page = new_doc.new_page(width=page.rect.width, height=page.rect.height)

            # تحويل الصفحة إلى صورة عالية الدقة
            # DPI (Dots Per Inch) - كلما زاد الرقم زادت الدقة
            pix = page.get_pixmap(dpi=300)
            img_bytes = pix.tobytes("png")
            pil_img = Image.open(io.BytesIO(img_bytes))

            # استخدام Tesseract OCR لاستخراج النص العربي
            print("جاري استخراج النص من الصورة...")
            # نستخدم 'ara' لتحديد اللغة العربية
            extracted_text = pytesseract.image_to_string(pil_img, lang='ara')

            if not extracted_text.strip():
                print("لم يتم العثور على أي نص في هذه الصفحة.")
                continue

            print("تم استخراج النص. جاري تصحيحه وكتابته في الملف الجديد...")

            # إصلاح النص المستخرج
            reshaped_text = arabic_reshaper.reshape(extracted_text)
            bidi_text = get_display(reshaped_text)

            # إضافة النص المصحح إلى الصفحة الجديدة
            # نستخدم textbox لوضع كل النص في مربع واحد
            rect = fitz.Rect(50, 50, page.rect.width - 50, page.rect.height - 50) # هوامش 50 بكسل
            new_page.insert_textbox(rect, bidi_text, fontfile=font_path, fontsize=12, align=fitz.TEXT_ALIGN_RIGHT)

        if len(new_doc) > 0:
            new_doc.save(output_pdf_path, garbage=4, deflate=True)
            print("\n" + "="*20)
            print(f"✅ تم الانتهاء بنجاح!")
            print(f"تم حفظ الملف المصحح في: {output_pdf_path}")
        else:
            print("\n⚠️ تحذير: لم يتم إنشاء ملف جديد لأنه لم يتم العثور على أي نصوص في الملف الأصلي.")

    except Exception as e:
        print(f"❌ حدث خطأ أثناء المعالجة: {e}")
    finally:
        if 'doc' in locals() and not doc.is_closed:
            doc.close()
        if 'new_doc' in locals() and not new_doc.is_closed:
            new_doc.close()

# --- كيفية الاستخدام ---
# ==============================================================================
input_file = "/content/PDFMathTranslate/translate.cli.text.with.figure-dual.pdf"
# ==============================================================================

if not os.path.exists(input_file):
    print(f"خطأ: ملف الإدخال '{input_file}' غير موجود.")
else:
    output_file = os.path.splitext(input_file)[0] + "_corrected_with_ocr.pdf"
    fix_pdf_with_ocr(input_file, output_file, ARABIC_FONT_PATH)
```

```python
import arabic_reshaper
from bidi.algorithm import get_display
from pypdf import PdfReader
from reportlab.pdfgen import canvas
from reportlab.lib.pagesizes import letter
from reportlab.pdfbase import pdfmetrics
from reportlab.pdfbase.ttfonts import TTFont
from reportlab.lib.styles import ParagraphStyle
from reportlab.platypus import Paragraph, Frame
from reportlab.rl_config import defaultPageSize

def correct_arabic(text: str) -> str:
    """إعادة تشكيل النص العربي وضبط اتجاهه."""
    reshaped = arabic_reshaper.reshape(text)
    bidi_text = get_display(reshaped)
    return bidi_text

def extract_arabic_from_pdf(path_pdf: str, page_number: int = 0) -> str:
    """استخراج النص من صفحة معينة (قد يكون معكوسًا)."""
    reader = PdfReader(path_pdf)
    page = reader.pages[page_number]
    txt = page.extract_text()
    return txt

def create_new_pdf_with_corrected_arabic(out_path: str, corrected_text: str):
    """إنشاء ملف PDF جديد يحتوي النص العربي بعد تصحيحه."""
    c = canvas.Canvas(out_path, pagesize=letter)
    # تسجيل خط يدعم العربية
    arabic_font = "Amiri"  # يمكنك استخدام أي خط عربي مناسب
    pdfmetrics.registerFont(TTFont(arabic_font, "/content/PDFMathTranslate/Amiri-Regular.ttf"))
    # إعداد نمط الفقرة يدعم RTL
    style = ParagraphStyle(
        name="ArabicStyle",
        fontName=arabic_font,
        fontSize=14,
        leading=18,
        alignment=2,  # 2 تعني محاذاة يمين في ReportLab
        rightIndent=0,
        leftIndent=0,
        wordWrap="RTL"
    )
    # استخدم إطار لوضع النص داخل صفحة
    frame = Frame(50, 100, defaultPageSize[0] - 100, defaultPageSize[1] - 200, showBoundary=0)
    para = Paragraph(corrected_text, style)
    frame.addFromList([para], c)
    c.showPage()
    c.save()

def main():
    input_pdf = "/content/PDFMathTranslate/translate.cli.text.with.figure-mono.pdf"
    output_pdf = "output_corrected.pdf"
    # استخراج النص (قد يكون معكوسًا)
    original = extract_arabic_from_pdf(input_pdf, page_number=0)
    print("Original extracted:", original)
    # تصحيح النص العربي
    fixed = correct_arabic(original)
    print("Fixed text:", fixed)
    # إنشاء PDF جديد بالنص المصحَّح
    create_new_pdf_with_corrected_arabic(output_pdf, fixed)
    print("Saved corrected PDF:", output_pdf)

if __name__ == "__main__":
    main()
```

```python
!pip install arabic-reshaper python-bidi pypdf reportlab

from google.colab import files
uploaded = files.upload()  # ارفع ملف Amiri-Regular.ttf و/أو ملف PDF إن لزم الأمر

import arabic_reshaper
from bidi.algorithm import get_display
from pypdf import PdfReader
from reportlab.pdfgen import canvas
from reportlab.lib.pagesizes import letter
from reportlab.pdfbase import pdfmetrics
from reportlab.pdfbase.ttfonts import TTFont

def correct_arabic(text: str) -> str:
    reshaped = arabic_reshaper.reshape(text)
    bidi_text = get_display(reshaped)
    return bidi_text

def read_arabic_text_from_pdf(input_pdf_path: str, page_number: int = 0) -> str:
    reader = PdfReader(input_pdf_path)
    page = reader.pages[page_number]
    txt = page.extract_text()
    return txt or ""

def create_new_pdf_with_corrected_arabic(out_path: str, corrected_text: str):
    c = canvas.Canvas(out_path, pagesize=letter)
    arabic_font = "Amiri"
    # تأكَّد أنك رفعت ملف الخط Amiri-Regular.ttf، واستخدم المسار الصحيح
    pdfmetrics.registerFont(TTFont(arabic_font, "/content/PDFMathTranslate/Amiri-Regular.ttf"))

    c.setFont(arabic_font, 20)
    # نحدد نقطة النهاية من اليمين (الهامش)
    x = letter[0] - 50  # العرض الكلي minus هامش 50
    y = 700

    c.drawRightString(x, y, corrected_text)

    c.showPage()
    c.save()

def main():
    input_pdf = "/content/PDFMathTranslate/output.pdf"
    output_pdf = "/content/PDFMathTranslate/output.pdf"

    # اقرأ النص من الصفحة 0 (أو أي صفحة تريدها)
    original_text = read_arabic_text_from_pdf(input_pdf, page_number=0)
    print("Original extracted:", original_text)

    # صحح النص العربي
    fixed_text = correct_arabic(original_text)
    print("Fixed (reshaped & bidi):", fixed_text)

    # أنشئ PDF جديد بالنص المصحَّح
    create_new_pdf_with_corrected_arabic(output_pdf, fixed_text)
    print("Saved corrected PDF to:", output_pdf)

    # تنزيل الملف الناتج
    files.download(output_pdf)

if __name__ == "__main__":
    main()
```

```python
from PyPDF2 import PdfReader, PdfWriter
from reportlab.pdfgen import canvas
from reportlab.lib.pagesizes import letter
from reportlab.pdfbase import pdfmetrics
from reportlab.pdfbase.ttfonts import TTFont
import arabic_reshaper
from bidi.algorithm import get_display
import io

# ملف PDF الأصلي
input_pdf = "/content/PDFMathTranslate/translate.cli.text.with.figure-dual (6).pdf"
output_pdf = "output_modified.pdf"

# النص العربي الجديد
arabic_text = "مرحبا بالعالم (تم التعديل)"
reshaped_text = arabic_reshaper.reshape(arabic_text)
bidi_text = get_display(reshaped_text)

# إنشاء صفحة PDF مؤقتة بالنص الجديد
packet = io.BytesIO()
c = canvas.Canvas(packet, pagesize=letter)
pdfmetrics.registerFont(TTFont('ArabicFont', "/content/PDFMathTranslate/Amiri-Regular.ttf"))
c.setFont("ArabicFont", 20)
c.drawRightString(550, 700, bidi_text)  # محاذاة يمين
c.save()

# إعادة المؤشر للبداية
packet.seek(0)

# دمج الصفحة الجديدة مع الملف الأصلي
new_pdf = PdfReader(packet)
existing_pdf = PdfReader(open(input_pdf, "rb"))
output = PdfWriter()

# نأخذ الصفحة الأولى كمثال ونضيف النص فوقها
page = existing_pdf.pages[0]
page.merge_page(new_pdf.pages[0])
output.add_page(page)

# إضافة باقي الصفحات كما هي
for i in range(1, len(existing_pdf.pages)):
    output.add_page(existing_pdf.pages[i])

# حفظ الملف المعدل
with open(output_pdf, "wb") as f:
    output.write(f)

print(f"✅ تم تعديل ملف PDF وحفظه في: {output_pdf}")
```

```python
from pypdf import PdfReader
from reportlab.pdfgen import canvas
from reportlab.lib.pagesizes import letter
from reportlab.pdfbase import pdfmetrics
from reportlab.pdfbase.ttfonts import TTFont
import arabic_reshaper
from bidi.algorithm import get_display

# مسار ملف PDF الأصلي
input_pdf = "/content/PDFMathTranslate/translate.cli.text.with.figure-dual (6) (1).pdf"
# ملف الإخراج
output_pdf = "reversed_arabic.pdf"

# تحميل الخط العربي
ARABIC_FONT_PATH = "/content/PDFMathTranslate/Amiri-Regular.ttf"
pdfmetrics.registerFont(TTFont("ArabicFont", ARABIC_FONT_PATH))

# قراءة النصوص من PDF
reader = PdfReader(input_pdf)

# إنشاء ملف PDF جديد
c = canvas.Canvas(output_pdf, pagesize=letter)
c.setFont("ArabicFont", 16)

y = 750  # إحداثي البداية للكتابة

for page in reader.pages:
    text = page.extract_text()
    if text:
        # إعادة تشكيل النص العربي
        reshaped = arabic_reshaper.reshape(text)
        bidi_text = get_display(reshaped)

        # عكس النص (حروف معكوسة)
        reversed_text = bidi_text[::-1]

        # كتابة النص المعكوس
        c.drawString(100, y, reversed_text)
        y -= 30  # نزول للسطر التالي

c.save()
print(f"✅ تم إنشاء ملف PDF جديد مع النصوص العربية معكوسة: {output_pdf}")
```

```python
import fitz  # PyMuPDF
import pytesseract
from PIL import Image
import arabic_reshaper
from bidi.algorithm import get_display
from reportlab.pdfgen import canvas
from reportlab.lib.pagesizes import letter
from reportlab.pdfbase import pdfmetrics
from reportlab.pdfbase.ttfonts import TTFont
import io

# ملف PDF الأصلي
input_pdf = "1.pdf"
# ملف الإخراج
output_pdf = "arabic_fixed.pdf"

# تحميل خط عربي
ARABIC_FONT_PATH = "Amiri-Regular.ttf"
pdfmetrics.registerFont(TTFont("ArabicFont", ARABIC_FONT_PATH))

doc = fitz.open(input_pdf)

# إنشاء ملف PDF جديد
c = canvas.Canvas(output_pdf, pagesize=letter)
c.setFont("ArabicFont", 14)

y = 750

for page_num in range(len(doc)):
    page = doc[page_num]
    # محاولة استخراج النص مباشرة
    text = page.get_text("text")

    if not text.strip():
        # لو النص غير موجود → OCR
        pix = page.get_pixmap()
        img = Image.frombytes("RGB", [pix.width, pix.height], pix.samples)
        text = pytesseract.image_to_string(img, lang="ara")

    if text.strip():
        # إعادة تشكيل وضبط الاتجاه
        reshaped = arabic_reshaper.reshape(text)
        bidi_text = get_display(reshaped)

        # كتابة النص المصحح
        for line in bidi_text.split("\n"):
            if line.strip():
                c.drawRightString(550, y, line.strip())
                y -= 20
        y -= 10  # مسافة إضافية بين الفقرات

c.save()
print(f"✅ تم إنشاء ملف PDF جديد بالنص العربي المصحح: {output_pdf}")
```
